```python
import pandas as pd

df = pd.read_csv("vendas_pais.csv", sep=";")

display(df)
```

```python
maior_venda_por_pais = {}

for i in range(len(df)):

    if df["pais"][i] in maior_venda_por_pais:
        continue
    else:
        maior_venda_por_pais[ df["pais"][i] ] = [ df["id_produto"][i], df["volumes"][i]]
```

```python
for pais in maior_venda_por_pais:
    print(f" - País: {pais[:-1]}\n - ID do produto mais vendido: {maior_venda_por_pais[pais][0]}\n - Quantidade vendida: {maior_venda_por_pais[pais][1]}\n")
```
