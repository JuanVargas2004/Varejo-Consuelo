#Editando JSON - *vendas_online.json*

```python
dataframe = pd.read_json('/content/drive/MyDrive/Colab_Notebooks/Projeto_Consuelo/Landing/vendas_online.json')

display(dataframe)
```

```python
#encontrando a linha onde começa a separação
for i in range(len(dataframe)):
  if np.isnan(dataframe["idVenda"][i]):
    separador =  i
    print(i)
    break
```

```python
#printando as colunas
print(dataframe.columns)
```

```python
#subtituir as colunas da tabela
for i in range(0, separador):
  dataframe.loc[i, 'ID_Venda'] = dataframe['idVenda'][i]
  dataframe.loc[i, 'ID_Produto'] = dataframe['idProduto'][i]
  dataframe.loc[i, 'ID_Loja'] = dataframe['idLoja'][i]
  dataframe.loc[i, 'Quantidade'] = dataframe['qtd'][i]
  dataframe.loc[i, "Data_Venda"] = dataframe['dataVenda'][i]
  dataframe.loc[i, "Preco_Unitario"] = dataframe['precoUnit'][i]
  dataframe.loc[i, "ID_Cliente"] = dataframe['idCliente'][i]
  dataframe.loc[i, "Desconto"] = dataframe['desconto'][i]
  dataframe.loc[i, "Promocao"] = dataframe['promocao'][i]
  dataframe.loc[i, "Estoque_Inicial"] = dataframe['estoqueInicial'][i]
  dataframe.loc[i, "Estoque_Final"] = dataframe['estoqueFinal'][i]
  dataframe.loc[i, "Devolucao"] = dataframe['devolucao'][i]
  dataframe.loc[i, "Canal_Venda"] = dataframe['canalVenda'][i]
```

```python
display(dataframe)
```

```python
#criando uma cópia do dataframe

final_dataframe = dataframe.iloc[:, 13:]

display(final_dataframe)
```

```python
#arredondando os valores

final_dataframe['Preco_Unitario'] = final_dataframe["Preco_Unitario"]
final_dataframe['Desconto'] = final_dataframe["Desconto"]

display(final_dataframe)
```

```python
#transformando as datas para o formato ano-mês-dia
final_dataframe["Data_Venda"] = pd.to_datetime(final_dataframe["Data_Venda"])
final_dataframe["Data_Venda"] = final_dataframe["Data_Venda"].dt.date

display(final_dataframe)
```

```python
#salvando arquivos

#final_dataframe.to_json('/content/drive/MyDrive/Colab_Notebooks/Projeto_Consuelo/Arquivos_Tratados/vendas_online.json', orient='records')


final_dataframe.to_csv('/content/drive/MyDrive/Colab_Notebooks/Projeto_Consuelo/Arquivos_Tratados/vendas_online.csv', index=False, encoding='utf-8')
```

# EDITANDO CSV - *campanhas_marketing.csv*

```python
dataframe = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/Projeto_Consuelo/Landing/campanhas_marketing.csv')

display(dataframe)
```

```python
#transformando as datas para o formato ano-mês-dia

#transformando o tipo da coluna de str para datetime
dataframe["Data_Inicio"] = pd.to_datetime(dataframe["Data_Inicio"])
dataframe["Data_Fim"] = pd.to_datetime(dataframe["Data_Fim"])

#separando apenas os anos, mês e dias
dataframe["Data_Inicio"] = dataframe["Data_Inicio"].dt.date
dataframe["Data_Fim"] = dataframe["Data_Fim"].dt.date

display(dataframe)
```
