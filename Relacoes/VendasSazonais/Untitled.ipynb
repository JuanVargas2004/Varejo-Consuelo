```python
import mysql.connector
import pandas as pd

# Conexão ao banco de dados MySQL
conn = mysql.connector.connect(
    #host="",
    user='root',
    password='0000',
    database='varejo'
)

# Query para buscar os dados
query = '''
SELECT pais, estacao, v.id_produto, p.nome_produto, SUM(quantidade) AS volumes
FROM vendas_sazonais v
JOIN produtos p ON v.id_produto = p.id_produto
GROUP BY estacao, pais, v.id_produto, p.nome_produto
ORDER BY volumes DESC
'''

# Carregar dados em um DataFrame
df = pd.read_sql(query, conn)

# Fechar a conexão
conn.close()

# Função para obter os 5 produtos mais vendidos por país e estação
def top_5_produtos_por_pais_estacao(df):
    return df.groupby(['pais', 'estacao']).apply(lambda x: x.nlargest(1, 'volumes')).reset_index(drop=True)

# Obter os 5 produtos mais vendidos por país e estação
top_produtos = top_5_produtos_por_pais_estacao(df)

# Converter o resultado para string
resultado_texto = top_produtos.to_string(index=False)

# Exibir o resultado
print(resultado_texto)

# Se desejar copiar o resultado para a área de transferência (funciona apenas em alguns ambientes)
try:
    import pyperclip
    pyperclip.copy(resultado_texto)
    print("O resultado foi copiado para a área de transferência.")
except ImportError:
    print("A biblioteca 'pyperclip' não está instalada. Instale-a para copiar automaticamente para a área de transferência.")

# Código para instalar a biblioteca 'pyperclip' se necessário:
# pip install pyperclip
```
