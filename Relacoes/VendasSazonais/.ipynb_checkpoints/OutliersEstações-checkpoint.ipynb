```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df = pd.read_csv("VendasSazonais.csv", sep=";")

display(df)
```

```python
import matplotlib.pyplot as plt

for i in range(len(df)):
    linha = df.iloc[i, 1:]

    fig = plt.figure(figsize=(10, 7))
    box = plt.boxplot(linha, patch_artist=True)

    upper_limit = box['whiskers'][1].get_ydata()[1]
    
    outliers = [flier.get_ydata() for flier in box['fliers']]
    upper_outliers = [flier for flier in outliers if len(flier) > 0 and any(value > upper_limit for value in flier)]

    if upper_outliers:
        print(f"Índice da linha: {i + 1}")
        print(f"Limite máximo: {upper_limit}")
        print("Outliers acima do limite máximo:")
        for flier in upper_outliers:
            print(flier)
        print("\n")

    plt.close(fig)
```
