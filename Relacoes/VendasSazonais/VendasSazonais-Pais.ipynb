# Importando tabela

```python
import pandas as pd

df = pd.read_csv("VendasGeral.csv", sep=";")

display(df)
```

# Padronizar Data

```python
def padronizar_data(data):
    final_data = data.split("-")

    ano, mes, dia = final_data

    if mes[0] == "0":
        mes = mes[1]

    if dia[0] == "0":
        dia = dia[1]

    return int(ano), int(mes), int(dia)

padronizar_data(df["data_venda"][0])
```

# Definir Estação

```python
def definir_estacao(pais, data):
    hemisferios_pt = {
        'Singapura': 'N',
        'Indonésia': 'S',
        'Chile': 'S',
        'Estados Unidos': 'N',
        'Israel': 'N',
        'África do Sul': 'S',
        'Malásia': 'N',
        'Jordânia': 'N',
        'Egito': 'N',
        'Hong Kong': 'N',
        'Índia': 'N',
        'Nigéria': 'N',
        'República Democrática do Congo': 'N',
        'França': 'N',
        'Sri Lanka': 'N',
        'Líbano': 'N',
        'Síria': 'N',
        'Emirados Árabes Unidos': 'N',
        'Japão': 'N',
        'Alemanha': 'N',
        'Senegal': 'N',
        'Tunísia': 'N',
        'Peru': 'S',
        'Itália': 'N',
        'Camboja': 'N',
        'Taiwan': 'N',
        'Turquia': 'N',
        'Brasil': 'S',
        'Vietnã': 'N',
        'Austrália': 'S',
        'Coreia do Sul': 'N',
        'Arábia Saudita': 'N',
        'Marrocos': 'N',
        'Rússia': 'N',
        'Catar': 'N',
        'Argélia': 'N',
        'Tailândia': 'N',
        'Argentina': 'S',
        'China': 'N',
        'Afeganistão': 'N',
        'Paquistão': 'N',
        'Filipinas': 'N',
        'Líbia': 'N',
        'Irã': 'N',
        'Mianmar': 'N',
        'Reino Unido': 'N',
        'Colômbia': 'N',
        'México': 'N',
        'Nepal': 'N',
        'Espanha': 'N',
        'Bangladesh': 'N',
        'Canadá': 'N',
        'Quênia': 'N',
        'Iraque': 'N'
    }

    #pegando o hemisfério
    hemisferio = hemisferios_pt[pais]
    

    #pegando o mes
    data_final = padronizar_data(data)
    ano, mes, dia = data_final

    if (mes >= 3 and mes < 6):
        estacao = "Outono" if hemisferio == 'S' else "Primavera"
    elif (mes >= 6 and mes < 9):
        estacao = "Inverno" if hemisferio == 'S' else "Verão"
    elif (mes >= 9 and mes < 12):
        estacao = "Primavera" if hemisferio == 'S' else "Outono"
    else:
        estacao = "Verão" if hemisferio == 'S' else "Inverno"

    return estacao
```

# Criando nova tabela

```python
# Aplicar a função e adicionar a coluna 'estacao'
df['estacao'] = df.apply(lambda row: definir_estacao(row['pais'][:-1], row['data_venda']), axis=1)

display(df)
```

```python
df.to_csv("VendasGeral-país.csv", index=False, sep=";")
```
