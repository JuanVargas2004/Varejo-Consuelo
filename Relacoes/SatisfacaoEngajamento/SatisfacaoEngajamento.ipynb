```python
import pandas as pd
import scipy.stats as stats


# Criando a tabela de contingência
contingency_table = pd.crosstab(df['tipo_engajamento'], df['satisfacao'])
print("Tabela de Contingência:\n", contingency_table)

# Realizando o teste qui-quadrado
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nEstatística Qui-Quadrado:", chi2)
print("\nValor-p:", p)
print("\nGraus de Liberdade:", dof)
print("\nFrequências Esperadas:\n", expected)
```
